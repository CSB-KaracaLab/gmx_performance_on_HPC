## DRAWING GPU GRAPH FROM OBTAINED RESULTS 

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt


gpu_for_barbuncuda=[0,1,2]
barbuncuda_results= [14.285,40.038, 40.012]


gpu_for_akyacuda=[0,1,2,3,4]
akyacuda_results=[14.924,71.372, 69.463, 69.360, 71.072]


fig, ax1=plt.subplots()

ax1.set_xlabel('GPU core number', fontname='Times New Roman', fontsize=12)
ax1.set_ylabel('Performance (ns/day)', fontname='Times New Roman', fontsize=12)
ax1.plot(gpu_for_barbuncuda,barbuncuda_results, color="tomato", label="          Barbun-cuda", marker="o")
ax1.plot()

ax1.plot(gpu_for_akyacuda,akyacuda_results, color="silver", label="          Akya-cuda", marker="o")
ax1.set_xticks(np.arange(0,5,1))
plt.legend()

ax2 = fig.add_axes([0.65,0.59,0.2,0.18])
ax3 = fig.add_axes([0.65,0.34,0.2,0.18])


ax2.plot(gpu_for_akyacuda[1:],akyacuda_results[1:], color="silver", marker="o", markersize=4)
ax2.set_xticks(np.arange(1,5,1))
ax2.set_yticks(np.arange(69,73,1))
ax3.plot(gpu_for_barbuncuda[1:],barbuncuda_results[1:], color="tomato", marker="o", markersize=4)
ax3.set_xticks(np.arange(1,5,1))
ax3.set_yticks(np.arange(39,42,1))
plt.savefig('GPU_performance_detailed_3.png', dpi=500, bbox_inches='tight')
```
